```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("purchase_history.csv")
```

```python
gender_encoded = pd.get_dummies(df['Gender'],drop_first=True)
```

```python
gender_encoded
```

```python
df = pd.concat([df,gender_encoded],axis=1)
```

```python
df
```

```python
df.drop(columns=['Customer ID','Product ID'],inplace=True)
```

```python
df.rename(columns={'Male': 'male'}, inplace=True)
df.rename(columns={'Age': 'age'}, inplace=True)
df.rename(columns={'Salary': 'salary'}, inplace=True)
df.rename(columns={'Price': 'price'}, inplace=True)
```

```python
df.to_csv("new_customers.csv", index=False)
```

```python
x = df[['male','age','salary','price']].to_numpy()
```

```python
x
```

```python
y = df['Purchased'].to_numpy()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k= 5
knn = KNeighborsClassifier(n_neighbors=k)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
